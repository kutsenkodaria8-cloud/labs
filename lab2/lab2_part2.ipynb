```python
import pandas as pd
import numpy as np
import timeit
```

```python
# завантаження датасету
# файл має роздільник ;, тому вказано sep=';' low_memory=False дозволяє коректно зчитати великий файл
```

```python
df = pd.read_csv('household_power_consumption.txt', sep=';', low_memory=False)
print(df.shape)
df.head()
```

```python
# очищення даних

# у датасеті пропущені значення позначені символом ?. замінюємо їх на nan.видаляємо всі рядки з пропущеними значеннями через dropna().
# створюємо нову колонку Datetime, об'єднавши колонки Date та Time
```

```python
df.replace('?', np.nan, inplace=True)
df.dropna(inplace=True)

df['Global_active_power'] = df['Global_active_power'].astype(float)
df['Global_intensity'] = df['Global_intensity'].astype(float)
df['Sub_metering_1'] = df['Sub_metering_1'].astype(float)
df['Sub_metering_2'] = df['Sub_metering_2'].astype(float)
df['Sub_metering_3'] = df['Sub_metering_3'].astype(float)

df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d/%m/%Y %H:%M:%S')

df.head()
```

```python
# потужність 5+ кВт
# вибірка всіх записів, у яких загальна споживана потужність 'Global_active_power' 5+ кВт. час виконання вимірюється за допомогою %timeit.
```

```python
def power(data):
    return data[data['Global_active_power'] > 5]

%timeit power(df)
power(df)
```

```python
#струм 19-20 А
# записи, де сила струму в межах 19-20 А 
# серед них додатково фільтруємо ті, у яких споживання пральна машина, холодильник перевищує споживання бойлера та кондиціонеру
#час виконання вимірюється за допомогою %timeit
```

```python
def filter_current(data):
    result = data[(data['Global_intensity'] >= 19) & (data['Global_intensity'] <= 20)]
    return result[result['Sub_metering_2'] > result['Sub_metering_3']]

%timeit filter_current(df)
filter_current(df)
```

```python
# 500 000 записів

# випадковим чином обираємо 500 000 записів без повторів (replace=False)

# для отриманої вибірки обчислюємо середні значення по всіх 3x групах споживання електрoенергії Sub_metering_1, Sub_metering_2, Sub_metering_3
```

```python
def random(data):
    sample = data.sample(n=500000, random_state=42)
    print(sample[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].mean())
    return sample

%timeit -n 3 random(df)
random(df)
```

```python
# споживання після 18:00

# обираємо записи після 18:00 з потужністю понад 6 кВт.  
# серед них залишаємо лише ті, де група Sub_metering_2 є найбільшою серед усіх трьох груп 
# ділимо результат на дві половини, з першої беремо кожен третій запис (починаючи з індексу 2), з другої кожен четвертий (починаючи з 3)
# об'єднуємо частини у результат
```

```python
def evening(data):
    evening = data[data['Datetime'].dt.hour >= 18]
    high = evening[evening['Global_active_power'] > 6]
    group2_biggest = high[(high['Sub_metering_2'] > high['Sub_metering_1']) & 
                          (high['Sub_metering_2'] > high['Sub_metering_3'])]
    
    half = len(group2_biggest) // 2
    first  = group2_biggest.iloc[:half].iloc[2::3]
    second = group2_biggest.iloc[half:].iloc[3::4]
    return pd.concat([first, second])

%timeit evening(df)
evening(df)
```

```python
#нормалізація приводить значення до діапазону від 0 до 1 за формулою (x - min)/(max - min)
#стандартизація приводить дані до розподілу з середнім 0 та стандартним відхиленням 1 за формулою (x - mean)/std
```

```python
cols = ['Global_active_power', 'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']

df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())

df_std = (df[cols] - df[cols].mean()) / df[cols].std()

print('Нормалізований:')
display(df_norm.head())
print('Стандартизований:')
display(df_std.head())
```

```python
# пірсон вимірює лінійну залежність між двома змінними, значення близьке до 1 означає сильну пряму лінійну залежність
#спірмен вимірює монотонну залежність
```

```python
from scipy.stats import pearsonr, spearmanr

print('Пірсон:', pearsonr(df['Global_active_power'], df['Global_intensity']))
print('Спірмен:', spearmanr(df['Global_active_power'], df['Global_intensity']))
```

```python
# One Hot Encoding

# створюємо новий категоріальний атрибут час доби, який визначається на основі години з колонки Datetime:
#ніч від 0 до 6 години
#ранок від 6 до 12 години
#день від 12 до 18 години
#вечір від 18 до 24 години
```

```python
df['time_of_day'] = pd.cut(df['Datetime'].dt.hour, 
                            bins=[-1, 6, 12, 18, 24], 
                            labels=['ніч', 'ранок', 'день', 'вечір'])

df_encoded = pd.get_dummies(df['time_of_day'], prefix='time')

display(df_encoded.head(10))
```
